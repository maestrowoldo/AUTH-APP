Observe que a coluna 'Filhos' possui algumas entradas inválidas do tipo 'NaN'.

**Importante!!!**

De maneira geral, dados faltantes devem ser considerados caso a caso. Muitas vezes podem na verade ter um significado específico no contexto da coleta dos dados.

Quando a informação de uma (ou mais) variáveis não estiver disponível (por exemplo: falta de resposta, rasura na marcação da resposta, ou simplesmente não foi possível coletar aquela informação para aquele indivíduo), esta informação deve ser marcada (ou indicada) com algum símbolo especial ou permanecer em branco, o que pode resultar em valores 'null' ou 'NaN' na importação dos dados em versão eletrônica.

Não é SEMPRE que esses casos devem ser preenchidos com 0 (zero), pois o zero não significa falta de informação, e sim uma quantidade que foi medida, mas foi verificada como nula.

No caso da coluna filhos, atribuir ou não valores 0 para valores desconhecidos 'NaN' pode, por exemplo, impactar o cálculo de uma média de filhos por funcionário.

Para fins didáticos faremos a substituição dos valores 'NaN' por 'zero', considerando por exemplo que o formulário que gerou estes dados cotinha uma opção de 'Não se aplica' ou 'Não tenho filhos', gerando valores não numérios que ao serem 'convertidos' geraram entradas 'NaN'.

```python
# CÉLULA 22
### Análise Bivariada

# <font color='red'>**Exercício:**</font> Modifique o exemplo abaixo para gerar gráficos comparativos entre as variáveis qualitativas **Est.Civil** e **Instrucao**...

# Criação da Tabela de Contingência (Frequência Cruzada)
tabela_cruzada = pd.crosstab(df['Est.civil'], df['Inst'])

# Usa Subplots para colocar 2 gráficos lado a lado
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
plt.subplots_adjust(wspace=0.3)

# 1. Gráfico de Barras Agrupadas
tabela_cruzada.plot.bar(ax=axes[0], rot=0)
axes[0].set_title("Instrução por Estado Civil (Agrupado)")
axes[0].set_xlabel("Estado Civil")
axes[0].set_ylabel("Contagem")
axes[0].grid(axis='y', linestyle='--')
axes[0].legend(title='Instrução')

# 2. Gráfico de Barras Empilhadas
tabela_cruzada.plot.bar(ax=axes[1], stacked=True, rot=0)
axes[1].set_title("Instrução por Estado Civil (Empilhado)")
axes[1].set_xlabel("Estado Civil")
axes[1].set_ylabel("Contagem")
axes[1].grid(axis='y', linestyle='--')
axes[1].legend(title='Instrução')

plt.show()
```

<font color='red'>**Exercício:**</font>  Qual a média de idade dos funcionários?

```python
# CÉLULA 24
## Estatísticas Descritivas

# A partir destes métodos, responda as perguntas a seguir:

# <font color='red'>**Exercício:**</font> Qual a média de idade dos funcionários?

media_idade = df['Idade'].mean()
print(f"A média de idade dos funcionários é: {media_idade:.2f} anos")
```

<font color='red'>**Exercício:**</font>  Dos funcionários que têm filhos, qual o número mais comum?

```python
# CÉLULA 25
# <font color='red'>**Exercício:**</font> Dos funcionários que têm filhos, qual o número mais comum?

# Filtra apenas os funcionários com Filhos > 0
filhos_maior_que_zero = df[df['Filhos'] > 0]

# O número mais comum é a moda (mode())
moda_filhos = filhos_maior_que_zero['Filhos'].mode()

print("O número mais comum de filhos (entre os que têm filhos) é:")
print(moda_filhos)
```

<font color='red'>**Exercício:**</font>  Qual a variância das idades?

```python
# CÉLULA 26
# <font color='red'>**Exercício:**</font> Qual a variância das idades?

variancia_idade = df['Idade'].var()
print(f"A variância das idades é: {variancia_idade:.2f}")
```

<font color='red'>**Exercício:**</font>  Qual a média de salário e idade das pessoas casadas?

```python
# CÉLULA 27
# <font color='red'>**Exercício:**</font> Qual a média de salário e idade das pessoas casadas?

# Filtra o DataFrame para funcionários casados
casados_df = df[df['Est.civil'] == 'casado']

# Calcula as médias
medias_casados = casados_df[['Salario', 'Idade']].mean()

print("Média de Salário e Idade das pessoas casadas:")
print(f"Salário Médio: {medias_casados['Salario']:.2f} salários mínimos")
print(f"Idade Média: {medias_casados['Idade']:.2f} anos")
```

<font color='red'>**Exercício:**</font> Qual o maior salário?

```python
# CÉLULA 28
# <font color='red'>**Exercício:**</font> Qual o maior salário?

maior_salario = df['Salario'].max()
print(f"O maior salário é: {maior_salario:.2f} salários mínimos")
```

<font color='red'>**Exercício:**</font> Qual o funcionário que possui maior salário (informe seus atributos)

```python
# CÉLULA 29
# <font color='red'>**Exercício:**</font> Qual o funcionário que possui maior salário (informe seus atributos)

# idxmax() retorna o índice (Funcionario) da linha com o valor máximo
funcionario_max_salario = df['Salario'].idxmax()

# Usa .loc[] para obter todos os atributos desse funcionário
atributos_max_salario = df.loc[funcionario_max_salario]

print(f"O Funcionário com maior salário (índice {funcionario_max_salario}) possui os seguintes atributos:")
print(atributos_max_salario)
```
